```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/Web_scraping"
html = urlopen(url)
```

```python
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
title = soup.title
print(title)
```

```python
text = soup.get_text()
```

```python
#Print all the lines with html tags a,  <a> </a>
# The a tag represent hyperlinks, links to other pages
all_links = soup.find_all("a")
all_links
```

```python
#Print all the lines with hmtl tag table <table> </table>
all_tables = soup.find_all('table')
all_tables
```

```python
#Print all the lines with html tag tr which is coloumn row
rows = soup.find_all('tr')
print(rows[:8])
```

```python
#Print all the rows with html tag td
for row in rows:
    row_td = row.find_all('td')
    print(row_td)
type(row_td)
```

```python
# Exercise 1
hyper_links = soup.find_all('a', href=True)

#Print all the lines with tag a and attribute href
all_links = []
for links in hyper_links:
    all_links.append(links['href'])
print(f"all links: {all_links}")

#Print all the (working) links to extrernal web pages with tag a and attribute href
real_links = []
for links in hyper_links:
    element = links['href']
    if element[:8] == 'https://': 
        real_links.append(element)
print(f"links to other webpages: {real_links}")
```

```python
#Get all the info contained in table (with tag td) and put them in a list

import re

list_rows = []
for row in rows:
    cells = row.find_all('td')
    str_cells = str(cells)
    clean = re.compile('<.*?>')
    clean2 = (re.sub(clean, '',str_cells))
    list_rows.append(clean2)
print(list_rows)
```

```python
#Use the list to make a data frame
df = pd.DataFrame(list_rows)
df.head(10)
```

```python
#Clean and format the data frame
df1 = df[0].str.split(',', expand=True)
df1.head(10)
```

```python
#Clean and format the data frame
df1[0] = df1[0].str.strip('[')
df1.head(10)
```

```python
# Get info on the data frame
df1.info()
df1.shape
```

```python
#Exercise 2
# Let's do the same things we just did on a another website, a wiki page on golf

#Create an object soup containing html info on the page
url = "https://en.wikipedia.org/wiki/Golf"
html = urlopen(url)
mypage = BeautifulSoup(html, 'lxml')
```

```python
#Print the lines with hyper links, tag a and attribute href
hyper_links = mypage.find_all('a', href=True)

#Print all the lines with tag a and attribute href
all_links = []
for links in hyper_links:
    all_links.append(links['href'])
print(f"all links: {all_links}")

#Print all the (working) links to extrernal web pages with tag a and attribute href
real_links = []
for links in hyper_links:
    element = links['href']
    if element[:8] == 'https://': 
        real_links.append(element)
print(f"links to other webpages: {real_links}")
```

```python
element = mypage.find_all('tr')
```

```python
#Get all the info contained in table (with attribute td) and put them in a list

import re

list_rows = []
for row in element:
    cells = row.find_all('td')
    str_cells = str(cells)
    clean = re.compile('<.*?>')
    clean2 = (re.sub(clean, '',str_cells))
    list_rows.append(clean2)
print(list_rows)

# We can notice here than there are a lot of lines containing a td tag, there are a lot of different tables. 
# We need to find a clever way to seperate the cells from different tables
```

```python
# To seperate cells from different tables we need to iterate first over the lines with table tag and then get the cells info
tables = mypage.find_all('table')
list_rows = []

for num, table in enumerate(tables):
    cells = table.find_all('td')
    str_cells = str(cells)
    clean = re.compile('<.*?>')
    clean2 = (re.sub(clean, '',str_cells))
    list_rows.append(clean2)
print(list_rows)

#Each list represent a full table
```

```python
df = pd.DataFrame(list_rows)
df
```

```python
df1 = df[0].str.split(',', expand=True)
df1.head(10)
```

```python
df1[0] = df1[0].str.strip('[')
df1.head(10)
#Now we have all the information we need, each line represent the info in a table.
#The format is not very clear however. Although the info of each table can be found in each row, it is not obvious.
# Let's reconstruct the table of the line #2.
```

```python
score = list(df.iloc[1])
score = score[0].split(', ')
new_score = []
for i in score:
    new_score.append(i.strip('[').strip(']').strip('\n'))
new_score
```
